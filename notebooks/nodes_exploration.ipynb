```python
def parse_datetime(status):
    try:
        eval_result = ast.literal_eval(status)
    except:
        return None

    if 'created_at' in eval_result:
        return parser.parse(eval_result['created_at'])
    else:
        return None

nodes["last_tweet_time"] = nodes["status"].map(lambda s: parse_datetime(s))
```
